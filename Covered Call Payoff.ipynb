# 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

# 2. Call payoff

```python
def call_payoff(St, strike_price, premium):
    return np.where(St > strike_price, St - strike_price, 0) - premium
```

# 3. Define parameters

```python
# Bitcoin price 
spot_price = 17500 

# Short call
strike_price_short_call = 17500
premium_short_call = 500

# Bitcoin price range at expiration of the call
St = np.arange(0.9*spot_price,1.1*spot_price,1) 
```

# 4. Bitcoin spot payoff

```python
payoff_bitcoin_spot = St-17500
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_bitcoin_spot,label='Bitcoin Payoff',color='g')
plt.xlabel('Bitcoin Spot Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 5. Short 17500 strike call payoff

```python
payoff_short_call = call_payoff(St, strike_price_short_call, premium_short_call) * -1.0
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_short_call,label='Short 17500 Strike Call',color='r')
plt.xlabel('Bitcoin Spot Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 6. Covered call payoff

```python
covered_call_payoff = payoff_bitcoin_spot + payoff_short_call
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_bitcoin_spot,'--',label='Long Bitcoin Spot',color='g')
ax.plot(St,payoff_short_call,'--',label='Short 17500 Strike Call ',color='r')
ax.plot(St,covered_call_payoff,label='Covered Call')
plt.xlabel('Bitcoin Spot Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
